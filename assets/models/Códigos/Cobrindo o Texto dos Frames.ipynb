## Foi aplicada essa técnica somente nos védos que vieram da base V-Librasil

Para o Articulador 1 - V-LiBrasil

```python
import cv2
import numpy as np
import pytesseract
import os

# Caminho do Tesseract (ajuste se necessário)
pytesseract.pytesseract.tesseract_cmd = r"C:\Program Files\Tesseract-OCR\tesseract.exe"

# Caminho raiz de entrada e saída
entrada = r"C:\Users\Ligia\Downloads\Frames - Articulador1"
saida = r"C:\Users\Ligia\Downloads\Frames - Articulador1"

# Percorre todas as subpastas e arquivos
for pasta_atual, subpastas, arquivos in os.walk(entrada):
    for nome_arquivo in arquivos:
        if nome_arquivo.lower().endswith((".png", ".jpg", ".jpeg")):
            caminho_completo = os.path.join(pasta_atual, nome_arquivo)

            # Lê imagem
            img = cv2.imread(caminho_completo)
            if img is None:
                print(f"Erro ao ler: {caminho_completo}")
                continue

            # Convertendo para cinza para cinza
            gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

            # OCR com localização
            data = pytesseract.image_to_data(gray, output_type=pytesseract.Output.DICT)

            # Cobre cada texto detectado com um retângulo preto
            for i in range(len(data['text'])):
                texto = data['text'][i].strip()
                if texto:  # só tapa se achar texto
                    x = data['left'][i]
                    y = data['top'][i]
                    w = data['width'][i]
                    h = data['height'][i]
                    cv2.rectangle(img, (x, y), (x + w, y + h), (0, 0, 0), -1)

            # Define caminho de saída equivalente
            caminho_saida = os.path.join(saida, os.path.relpath(caminho_completo, entrada))
            os.makedirs(os.path.dirname(caminho_saida), exist_ok=True)

            # Salva imagem mascarada
            cv2.imwrite(caminho_saida, img)
            print(f"Salvo: {caminho_saida}")
```
